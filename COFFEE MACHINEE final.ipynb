```python

money = 550
water = 400
milk = 540
coffee_beans = 120
dis_cups = 9

def start():
    print(f"""
The coffee machine has:
{water} of water
{milk} of milk
{coffee_beans} of coffee beans
{dis_cups} of disposable cups
${money} of money
""")

def func():
    command = input("Write action ('buy','fill','take','remaining','exit'):")
    while True:
        if command =='buy':
            buy()
            break
            
        elif command == 'fill':
            fill()
            break
            
        elif command == 'take':
            take()
            break
            
        elif command == 'remaining':
            remaining()
            break
        else:
            exit()
            
    

    

def buy():
    global money,water,milk,coffee_beans,dis_cups

    command_buy = input("What do you want to buy? 1 - espresso, 2 - latte, 3 - cappuccino,back- to main menu:")
    if command_buy == '1':
        
        if water < 250:
            print('Sorry, not enough water!')
            func()
        if coffee_beans < 16:
            print('Sorry, not enough coffee beans!')
            func()
        if dis_cups < 1:
            print('Sorry, not enough cups!')
            func()
        else:  
            water = water -250
            coffee_beans = coffee_beans - 16
            money = money + 4
            dis_cups = dis_cups -1
            print("I have enough resources, making you a coffee!")
            func()
            
       
        
    
    elif command_buy == '2':
        
        if water < 350:
            print('Sorry, not enough water!')
            func()
        if milk < 75:
            print('Sorry, not enough milk!')
            func()
        if coffee_beans < 20:
            print('Sorry, not enough coffee beans!')
            func()
        if dis_cups < 1:
            print('Sorry, not enough cups!')
            func()
        else:

            water = water - 350
            milk = milk - 75
            coffee_beans = coffee_beans - 20
            money = money + 7
            dis_cups = dis_cups -1
            print("I have enough resources, making you a coffee!")
            func()
        
     
    
    elif command_buy == '3':
        
        if water < 200:
            print('Sorry, not enough water!')
            func()
        if milk < 100:
            print('Sorry, not enough milk!')
            func()
        if coffee_beans < 12:
            print('Sorry, not enough coffee beans!')
            func()
        if dis_cups < 1:
            print('Sorry, not enough cups!')
            func()
        else:
            water = water - 200
            milk = milk - 100
            coffee_beans = coffee_beans - 12
            money = money + 6
            dis_cups = dis_cups -1
            print('I have enough resources, making you a coffee!')
            func()
        
    elif command_buy =='back':
        pass
    
    
def fill():
    global water, milk, coffee_beans, dis_cups
    water += int(input('Write how many ml of water do you want to add:'))
    milk += int(input('Write how many ml of milk do you want to add:'))
    coffee_beans += int(input('Write how many grams of coffee beans do you want to add:'))
    dis_cups += int(input('Write how many disposable cups of coffee do you want to add:'))
    func()

    
def take():
    global money
    print(f'I gave you ${money}')
    money -= money
    func()

def remaining():
    
    print(f"""
The coffee machine has:
{water} of water
{milk} of milk
{coffee_beans} of coffee beans
{dis_cups} of disposable cups
${money} of money
""")
    func()
    

func()

    
start()
```
